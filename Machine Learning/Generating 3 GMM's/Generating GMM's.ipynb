```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
```

4a)

```python
mu1 = np.array([0,0])
cov1 = 2*np.array([[1,0],[0,1]])
mu2 = np.array([5,1])
cov2 = 1*np.array([[6,0],[0,6]])
mu3 = np.array([-5,2])
cov3 = 1*np.array([[25,0],[0,25]])

X1 = np.random.multivariate_normal(mu1,cov1,100)
X2 = np.random.multivariate_normal(mu2,cov2,100)
X3 = np.random.multivariate_normal(mu3,cov3,100)
X = np.concatenate([X1,X2,X3])
```

```python
plt.scatter(X1[:,0],X1[:,1])
plt.scatter(X2[:,0],X2[:,1])
plt.scatter(X3[:,0],X3[:,1])
```

```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
kmeans = KMeans(n_clusters=3)
labels = kmeans.fit(X).predict(X)
plt.scatter(X[:,0],X[:,1],c=labels)
```

```python
g1 = GaussianMixture(n_components=3,covariance_type='spherical')
labels = g1.fit(X).predict(X)
plt.scatter(X[:,0],X[:,1],c=labels)
```
