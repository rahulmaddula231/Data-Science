```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os
from PIL import Image
import sys
plt.rcParams['figure.figsize']=[16,8]
```

```python
A=imread('mandrill_color.png')
x=np.mean(A,-1)
img=plt.imshow(x)
img.set_cmap('gray')
plt.savefig('gray.png')
plt.show()
```

```python
print(x.shape)
```

```python
imageheight=288
imagewidth=288
```

```python
u,s,vt=np.linalg.svd(x,full_matrices=True)
s=np.diag(s)
print(s)
```

```python
for i in range (0,len(s)):
  s[i] = sorted(s[i], reverse=True)
```

```python
plt.plot(s)
plt.show()
```

```python
j = []
for i in range (0,len(s)):
  j.append(s[i][0])
```

```python
plt.plot(j)
```

```python
u,s,vt=np.linalg.svd(x,full_matrices=True)
s=np.diag(s)
k=10
app=u[:,:10] @ s[0:10,:10] @vt[:10,:]
imgage1=plt.imshow(app)
imgage1.set_cmap('gray')
plt.show()
originalsize=imageheight*imagewidth
compressedsize=288*k*2+k*k
print(originalsize)
print(compressedsize)
ratio=compressedsize*1.0/originalsize
print(ratio)
```

```python
a = []
b = []
```

```python
u,s,vt=np.linalg.svd(x,full_matrices=True)
s=np.diag(s)
k=20
app=u[:,:20] @ s[0:20,:20] @vt[:20,:]
imgage1=plt.imshow(app)
imgage1.set_cmap('gray')
plt.show()

compressedsize1=288*k*2+k*k
print(originalsize)
print(compressedsize1)
ratio1=compressedsize1*1.0/originalsize
print(ratio1)
a.append(k)
b.append(ratio1)

```

```python
u,s,vt=np.linalg.svd(x,full_matrices=True)
s=np.diag(s)
k=40
app=u[:,:k] @ s[0:k,:k] @vt[:k,:]
imgage1=plt.imshow(app)
imgage1.set_cmap('gray')
plt.show()
compressedsize2=288*k*2+k*k
print(originalsize)
print(compressedsize2)
ratio2=compressedsize2*1.0/originalsize
print(ratio2)
a.append(k)
b.append(ratio2)
```

```python
u,s,vt=np.linalg.svd(x,full_matrices=True)
s=np.diag(s)
k=60
app=u[:,:k] @ s[0:k,:k] @vt[:k,:]
imgage1=plt.imshow(app)
imgage1.set_cmap('gray')
plt.show()
compressedsize3=288*k*2+k*k
print(originalsize)
print(compressedsize3)
ratio3=compressedsize3*1.0/originalsize
print(ratio3)
a.append(k)
b.append(ratio3)
```

```python
import pandas as pd

prediction = pd.DataFrame(b,a)

prediction
```

We can see that the image is almost clear, and that increasing k will provide a slightly greater image quality. However, increasing k would be pointless because the image is already read perfectly with k = 60, and the compression ratio is likewise good. This is due to the fact that utilizing k with 60 singular values usually always produces the same image matrix.
