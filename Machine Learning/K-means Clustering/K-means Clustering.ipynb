5b) We are implementing K-means clustering algorithm

```python
def k_means(X,k):
  m,n = X.shape
  centroids = np.random.rand(k,n)
  prev_centroids = np.array(np.random.randint(5,15,(k,n)))
  points = np.array([0]*m)
  while(np.sum(centroids-prev_centroids)**2>0.001):
    counts = np.array([0]*k)
    vec = np.zeros((k,n))
    for i in range(m):
      points[i] = np.argmin(np.sum((centroids-X[i])**2,axis=-1))
      counts[points[i]]+=1
      vec[points[i]]+=X[i]
    if(np.all(counts)):
      prev_centroids = centroids.copy()
      centroids = vec/counts.reshape(-1,1)
    else:
      centroids = np.random.rand(k,n)
  return points,centroids
```

k=3

```python
P,y = k_means(X,3)
```

```python
plt.scatter(X[:,0],X[:,1],c=P)
plt.scatter(y[:,0],y[:,1],c="red")
```
